```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
```

```python
df=yf.download("TCS.NS")
df
```

```python
df1=df[["High","Volume","Close"]]
df1
```

```python
df1.to_excel("ymcatcs.xlsx")
```

```python
#figsize(length , width) of graph
df1.plot(subplots=True,figsize=(8,8), marker="*",markersize=10,linewidth=5,linestyle="--")
plt.title("TCS report")
plt.xlabel("Date")
plt.ylabel("Sales in RS")
plt.show()
```

```python
df1.plot(kind='bar',color=["red","blue","green"])
plt.title("TCS report")
plt.xlabel("Date")
plt.ylabel("Sales in RS")
plt.show()
```

```python
df1.plot(kind='barh',color=["red","blue","green"])
plt.title("TCS report")
plt.xlabel("Date")
plt.ylabel("Sales in RS")
plt.show()
```

```python
df1.plot(kind='hist',color=["red","blue","green"])
plt.title("TCS report")
plt.xlabel("Date")
plt.ylabel("Sales in RS")
plt.show()
```

```python
df1.plot(kind='box', color=dict(boxes='r', whiskers='b', medians='g', caps='black'))
plt.title("TCS report")
plt.xlabel("Date")
plt.ylabel("Sales in RS")
plt.show()
```

```python
df1.plot(kind='area',color=["red","blue","green"])
plt.title("TCS report")
plt.xlabel("Date")
plt.ylabel("Sales in RS")
plt.show()
```

```python
df1.plot(kind='pie',color=["red","blue","green"], subplots=True)
plt.title("TCS report")
plt.xlabel("Date")
plt.ylabel("Sales in RS")
plt.show()
```

```python
df1.plot(kind='scatter', x='High', y='Volume', color="red")
plt.title("TCS report")
plt.xlabel("High Price")
plt.ylabel("Volume")
plt.show()
```
